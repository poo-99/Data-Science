## Estimation And Confidence Intervals

```python
# Import Libraries
import numpy as np
from scipy import stats
```

```python
# Sample data
data = [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
print(data)
```

## A. 99% Confidence Interval Using Sample Standard Deviation (t-distribution)

### Mean

```python
sample_mean=np.mean(data)
print(sample_mean)
```

### Degree of Freedom

```python
n=len(data)
dof=n-1
print(dof)
```

### Standard Error

```python
from scipy.stats import sem
SE=sem(data)
print(SE)
```

### Find Critical t-value

```python
alpha=0.01  # for 99% CI
```

```python
# t critical value
t_crit = stats.t.ppf(1 - 0.01, dof)
print(t_crit)
```

```python
# Two-tailed test
t_critical_two_tailed=stats.t.ppf(1-alpha/2,dof)

#One-tailed test
t_critical_one_tailed=stats.t.ppf(1-alpha,dof)

print("T Critical Value (Two-Tailed):",t_critical_two_tailed)
print("T Critical Value (One-Tailed):",t_critical_one_tailed)
```

### Margin of error

```python
ME=t_crit*SE
print(ME)
```

### Confidence interval

```python
CI_lower = sample_mean - ME
CI_upper = sample_mean + ME

print(f"99% Confidence Interval:({CI_lower}, {CI_upper})")
```

## B. 99% Confidence Interval Using Known Population Standard Deviation (z-distribution)

### Standard Error

```python
sigma=0.2
SE=(sigma/np.sqrt(n))
print(SE)
```

### Critical z-value for 99% CI

```python
z_crit=stats.norm.ppf(1-alpha/2)
print(z_crit)
```

### Margin of error

```python
margin_error_z = z_crit * SE
print(margin_error_z)
```

### Confidence interval

```python
ci_z=(sample_mean-margin_error_z,sample_mean+margin_error_z)
```
