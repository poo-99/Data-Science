# Exercise 1: Check if a number is prime

```python

# Program to check if a number is prime

num = int(input("Enter the number: "))

if num > 1:
    # check for factors
    for i in range(2, num):
        if (num % i) == 0:
            print(num, "is not a prime number")
            break
    else:
        print(num, "is a prime number")
# if input number is less than or equal to 1, it is not prime
else:
    print(num, "is not a prime number")
```

# Exercise 2: Product of two random numbers

```python
import random

# Generate two random integers between 1 and 10
num1 = random.randint(1, 10)
num2 = random.randint(1, 10)

# Calculate the product
product = num1 * num2

# Print the numbers and their product
print(f"Random number 1: {num1}")
print(f"Random number 2: {num2}")
print(f"Product: {product}")
```

# Exercise 3: Squares of Even or Odd Numbers

```python
# Squares of even and odd numbers between 100 and 200

print("Squares of Even numbers (100–200):")
for i in range(100, 201):   # 201 so that 200 is included
    if i % 2 == 0:
        print(i, "->", i ** 2)

print("\nSquares of Odd numbers (100–200):")
for i in range(100, 201):
    if i % 2 != 0:
        print(i, "->", i ** 2)
```

# Exercise 4: Word Counter1

```python
# Program to count frequency of words in a given text

text = input("Enter a text: ")

# Convert to lowercase for uniform counting
text = text.lower()

# Split into words
words = text.split()

# Create dictionary to store word counts
word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

# Print word frequencies
print("\nWord Frequency:")
for word, count in word_count.items():
    print(word, ":", count)
```

# Exercise 5: Check for Palindrome

```python
# Palindrome check ignoring spaces and case

text = input("Enter a string: ")

# Convert to lowercase and remove spaces
cleaned = text.replace(" ", "").lower()

if cleaned == cleaned[::-1]:
    print(text, "is a palindrome")
else:
    print(text, "is not a palindrome")
```
