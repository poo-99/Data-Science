## Descriptive Analysis and Data Preprocessing on Sales & Discounts Dataset

```python
#Upload Dataset
data=pd.read_csv("sales_data_with_discounts.csv")
```

```python
#Top 5 
data.head()
```

```python
#Check null
data.isnull().sum()
```

```python
#Find Numerical Column
numerical_col=data.select_dtypes(include=["int64","float64"]).columns
print(numerical_col)
```

```python
#Describe Numerical Column
data[numerical_col].describe()
```

```python
# Mean of NUmerical Column
Mean=data.mean(numeric_only=True)
print("Mean of numerical Data:\n",Mean)
```

```python
# Median of Numerical Column
Median=data.median(numeric_only=True)
print("Meadian of numerical Data:\n",Median)
```

```python
# Mode of Numerical Column
Mode=data.mode(numeric_only=True).iloc[0]
print("Mode of numerical Data:\n",Mode)
```

```python
# Standard Deviation of Numerical Column
std_values = data.std(numeric_only=True)
print("Standard Deviation :\n",std_values)
```

# Data Visualization

## Histogram for Numerical Columns

```python
for col in numerical_col:
    plt.figure(figsize=(6,4))
    sns.histplot(data=data[col],kde=True,bins=20)
    plt.title(f"Frequency distribution of {col}")
    plt.show()
```

Volume: Right-skewed; most values are in the lower range.

Avg Price: Shows distinct price points, likely due to different product categories.

Total Sales Value and Net Sales Value: Slightly skewed right, indicating occasional high-value transactions.

Discount Rate (%): Fairly normally distributed.

Discount Amount: Has a long tail  some high-discount items present.

## Boxplot for Numerical Columns

```python
for col in numerical_col:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=data[col])
    plt.title(f"Box Plots of {col}")
    plt.show()
```

Volume & Sales Values: Several outliers on the higher end.

Discount Amount: Very wide range; significant variance in how discounts are applied.

Avg Price: Some products priced significantly higher than the rest (outliers).


# Bar Chart Analysis for Categorical Column

```python
# Find Categorical Columns
categorical_col=data.select_dtypes(include="object").columns
categorical_col
```

```python
for col in categorical_col:
    plt.figure(figsize=(10, 5))
    sns.countplot(y=data[col],order=data[col].value_counts().index)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

#  Standardization of Numerical Variables(Z-score)

- Standard deviation () = 1

This helps when features (columns) are measured on different scales (e.g., height in cm vs. weight in kg). Many machine learning algorithms (like logistic regression, KNN, SVM, PCA) work better when features are standardized.

```python
from sklearn.preprocessing import StandardScaler
```

```python
#Initialize scaler
scaler = StandardScaler()

#Fit & transform numeric columns
standardized_array = scaler.fit_transform(data[numerical_col])

#Convert back to Dataframe with same column names
standardized_df = pd.DataFrame(standardized_array, columns=numerical_col)
standardized_df.head()
```

```python
# Histograms of  standardization

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.histplot(data["Total Sales Value"],kde=True,bins=20)
plt.title("Before Standardization: Total sales value")
plt.show()

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.histplot(standardized_df["Total Sales Value"],kde=True,bins=20)
plt.title("After Standardization: Total sales value")

plt.show()
```

## One-hot Encoding for Categorical Columns

```python
from sklearn.preprocessing import OneHotEncoder

#Initialise encoder
encoder=OneHotEncoder(drop="first",sparse_output=False)

#Fit & Transform categorical columns
encoded=encoder.fit_transform(data[categorical_col])

#Convert back to DataFrame with new column names
encoded_data=pd.DataFrame(encoded,columns=encoder.get_feature_names_out(categorical_col))

#Combine encoded categorical + numeric columns
data_encoded=pd.concat([data.drop(columns=categorical_col),encoded_data],axis=1)

# preview transformed dataset
print(data_encoded.head())
```
